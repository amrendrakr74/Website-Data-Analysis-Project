```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df = df.drop(index=0).reset_index(drop=True) 
df.head()
```

```python
df.columns = ["Channel group", "DateHour", "Users", "Sessions", "Engaged Sessions", "Average engagement time per session", "Engaged sessions per user","Events per session",	"Engagement rate",	"Event count"]
```

```python
df.head()
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H", errors='coerce')
```

```python
df.head()
```

```python
numeric_cols = df.columns.drop(["Channel group", "DateHour"])

df [numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

df ["Hour"]= df ["DateHour"].dt.hour
```

```python
df.info()
```

```python
df.describe()
```

```python
#Sessions and user over Time
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))

df.groupby("DateHour") [["Sessions", "Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and User over Time")
plt.xlabel("Datehour")
plt.ylabel("Count")


plt.show()
```

```python
#2nd ques
# Channel with highest users
```

```python
Channel_users = df.groupby('Channel group')['Users'].sum().sort_values(ascending=False)

print(Channel_users)

# Plot
Channel_users.plot(kind='bar', figsize=(8,4), title="Users by Channel")
plt.show()
```

```python
#3rd ques
#Highest average engagement time
```

```python
engagement_time = df.groupby('Channel group')['Average engagement time per session'].mean().sort_values(ascending=False)

print(engagement_time)

engagement_time.plot(kind='bar', figsize=(10,5))
plt.title("AVG ENGAGEMENT TIME BY CHANNEL", fontsize=14, fontweight='bold')
plt.show()
```

```python
#4th ques
#Engagement rate distribution by channel
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=df, x="Channel group", y="Engagement rate", palette="coolwarm")

plt.title(" Engagement Rate Distribution by Channel")

plt.xticks(rotation=45)

plt.show()
```

```python
#5th ques
#Engaged vs Non-engaged sessions
```

```python

df['Non Engaged Sessions'] = df['Sessions'] - df['Engaged Sessions']

engaged_compare = df.groupby('Channel group')[['Engaged Sessions', 'Non Engaged Sessions']].sum()

print(engaged_compare)

engaged_compare.plot(kind='bar', figsize=(10,5))
plt.title("ENGAGED vs NON-ENGAGED SESSIONS", fontsize=14, fontweight='bold')
plt.show()
```

```python
#6th ques
#Traffic by hour per channel
```

```python
heatmap_data = df.groupby(["Hour", "Channel group"]) ["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12, 6))

sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=5, annot=True, fmt='.0f')

plt.title(" Traffic by Hour and Channel")

plt.xlabel("Channel group")

plt.ylabel("Hour of Day")

plt.show()
```

```python
#7th ques
# Correlation between sessions & engagement rate
```

```python

df_plot =df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10, 5))

plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color="green")

plt.plot(df_plot["DateHour"], df_plot ["Sessions"], label="Sessions", color="blue")

plt.title(" Engagement Rate vs Sessions Over Time")

plt.xlabel("DateHour")

plt.legend()
```
